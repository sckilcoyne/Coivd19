```python
API = 'api.census.gov/data/2019/pep/population'

getData = 'POP,REGION,GEO_ID,COUSUB,CBSA,CSA,DIVISION,GEOCOMP,METDIV,NATION,PLACE,UNIVERSE'

callGet = '?get=' + getData

predicate = 'state'
predicateValue = '*'
callPredicate = '&for=' + predicate + ':' + predicateValue

callAPI = 'https://' + API + callGet + callPredicate
print(callAPI)
```

```python
censusDataRaw = requests.get(callAPI)
print('API call: ' + str(censusDataRaw.status_code) + ' ' + censusDataRaw.reason)
popData = censusDataRaw.json()
dfPop = pd.DataFrame(popData[1:], columns=popData[0])
dfPop = dfPop.set_index('state')
dfPop.index.names = ['FIPS']
dfPop.head(5)
```
